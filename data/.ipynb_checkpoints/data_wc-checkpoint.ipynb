```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
palette ={"A": "C0", "B": "C1"}
hue_order = ['A', 'B']
```

# World cups 1982-2018

In the Zone column, I divided the goal into 9 different sections looking at it from the front:

1. Upper left corner
2. Upper center
3. Upper right corner
4. Middle left area
5. Middle center area
6. Middle right area
7. Bottom left corner
8. Bottom center zone
9. Bottom right corner

```python
def a_b(df):
    a_b = []
    for i in df.index:
        if df.loc[i,'Started'] == 1:
            a_b.append('A')
        else:
            a_b.append('B')
    return a_b
```

```python
wc = pd.read_excel('wc_data.xlsx')
# Add column to know first miss
wc.rename(columns = {'Team': 'NT'}, inplace = True)
wc['First_Miss'] = np.zeros(wc.shape[0])
wc.loc[wc['Goal'] == 0, 'First_Miss'] = wc.loc[wc['Goal'] == 0, 'Penalty_Number']
# Add column by row
wc['Round'] = np.ceil(wc['Penalty_Number']/2).astype(int)
# Add column for starting, Started = A, nonstarting = B
wc['Started'] = (wc['Penalty_Number'] % 2 != 0).astype(int)
wc['Team'] = a_b(wc)
# Drop columns withNa
wc = wc.dropna(axis = 0)
wc.head()
```

# Studying the per game

```python
# To address question of how many penalties 
wc_by_game = (
    wc.groupby(['Game_id', 'Team'])
    .agg(
        Result = pd.NamedAgg(column = 'Goal', aggfunc =  'sum'), # Get the result
        Penalty_Number = pd.NamedAgg(column = 'Penalty_Number', aggfunc = lambda x: int(np.ceil(x.max()/2))), # Number of pens taken
        First_Miss = pd.NamedAgg(column = 'First_Miss', aggfunc = lambda x: x[x != 0].min()), # When was the first miss
        Home = pd.NamedAgg(column = 'Home', aggfunc = 'max')
    )
)
wc_by_game['Result'] = wc_by_game['Result'].astype(int)
wc_by_game['First_Miss_Round'] = np.ceil(wc_by_game['First_Miss']/2) # Which round was the first miss
wc_by_game[['First_Miss', 'First_Miss_Round']] = wc_by_game[['First_Miss', 'First_Miss_Round']].fillna(0).astype(int)
wc_by_game.head()
```

```python
# Appending a column for winners
def appendWinHA(df):
    df_c = df.copy()
    df_c.loc[:,'Win'] = np.zeros(df.shape[0])
    df_c.loc[df['Result'] == df['Result'].max(), 'Win'] = 1
    if df_c.loc[(slice(None),'A'), 'Win'].values == 1:
        df_c['WinStarted'] = [1, 0]
    else:
        df_c['WinStarted'] = [0, 0]
    df_c['Home'] = df_c['Home'].map({1: 'Home', 0: 'Away'})
    return df_c
wc_by_game = wc_by_game.groupby('Game_id').apply(appendWinHA)
```

```python
wc_by_game.sort_values(by=['Game_id', 'Team'], ascending = 'False')
```

```python
wc_by_game.groupby('Team')['Win'].sum()
```

```python
wc_by_game.groupby('WinStarted').sum()
```

```python
sns.countplot(x = wc_by_game[wc_by_game['Win'] == 1].index.get_level_values(1).sort_values())
plt.title('Number of wins in penalty shootouts by order of teams')
plt.ylabel('Number of Wins')
plt.xlabel('Team')
plt.ylim(0,17)
plt.show()
```

```python
sns.countplot(hue = wc_by_game[wc_by_game['Win'] == 1].index.get_level_values(1), x = wc_by_game.loc[wc_by_game['Win'] == 1, 'Home'].values, palette = palette, hue_order = hue_order)
plt.title('Number of wins in penalty shootouts by order of teams')
plt.ylabel('Number of Wins')
plt.xlabel('Team')
plt.show()
```

```python
sns.countplot(x = wc_by_game['First_Miss_Round'], hue = wc_by_game.index.get_level_values(1), palette = palette, hue_order = hue_order)
plt.title('Distribution of first misses per round')
plt.ylabel('Count')
plt.xlabel('Round')
plt.xlim(0.5,5.5)
plt.show()
```

```python
sns.countplot(x = wc_by_game.loc[wc_by_game['Win'] == 1, 'Penalty_Number'],
              hue = wc_by_game[wc_by_game['Win'] == 1].index.get_level_values(1), palette = palette, hue_order = hue_order)
plt.title('Number of penalties by team')
plt.ylabel('Count')
plt.xlabel('Round')
plt.show()
```
