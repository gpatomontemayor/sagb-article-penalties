```python
fifawc = pd.read_json('data/matches/43/3.json')
fifawc.head()
```

```python
fifawc_mod = pd.concat([fifawc[['match_id', 'match_date', 'kick_off', 'home_score', 'away_score']], 
                       fifawc['home_team'].apply(pd.Series)[['home_team_id', 'home_team_name']], 
                       fifawc['away_team'].apply(pd.Series)[['away_team_id', 'away_team_name']],
                       fifawc['competition_stage'].apply(pd.Series).rename({'id': 'comp_id', 'name': 'comp_name'}, axis = 1),
                       fifawc['stadium'].apply(pd.Series)['country'].apply(pd.Series)['name'].rename({'name': 'country'}, axis = 1)],
          axis = 1)
fifawc_mod
```
